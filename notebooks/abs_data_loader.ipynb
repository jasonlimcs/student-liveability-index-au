```python
import pandas as pd
import os
```

```python
RAW_PATH = "../data/raw/abs/"

# G01 - Basic Demographics
g01 = pd.read_csv(os.path.join(RAW_PATH, "2021Census_G01_AUST_SA2.csv"))
print('g01:', g01.columns.tolist())
g01 = g01[['SA2_CODE_2021', 'Tot_P_P', 'Age_20_24_yr_P', 'Age_25_34_yr_P']]
g01['pct_youth_18_34'] = 100 * (g01['Age_20_24_yr_P'] + g01['Age_25_34_yr_P']) / g01['Tot_P_P']
g01 = g01[['SA2_CODE_2021', 'Tot_P_P', 'pct_youth_18_34']]
g01.columns = ['sa2_code', 'total_pop', 'pct_youth_18_34']
print('g01:', g01.columns.tolist())

g02 = pd.read_csv(os.path.join(RAW_PATH, "2021Census_G02_AUST_SA2.csv"))
g02 = g02[['SA2_CODE_2021', 'Median_age_persons', 'Median_tot_prsnl_inc_weekly', 'Median_rent_weekly']]
g02.columns = ['sa2_code', 'median_age', 'median_weekly_income', 'median_weekly_rent']
# print('g02:', g02.columns.tolist())

g15 = pd.read_csv(os.path.join(RAW_PATH, "2021Census_G15_AUST_SA2.csv"))
g15 = g15[['SA2_CODE_2021', 'Tert_Tot_Tert_P', 'Tot_P']]
g15['pct_students'] = 100 * g15['Tert_Tot_Tert_P'] / g15['Tot_P']
g15 = g15[['SA2_CODE_2021', 'pct_students']]
g15.columns = ['sa2_code', 'pct_students']

df = g01 \
    .merge(g02, on='sa2_code', how='left') \
    .merge(g15, on='sa2_code', how='left')

df['rent_to_income_ratio'] = df['median_weekly_rent'] / df['median_weekly_income']
df.to_csv("../data/processed/abs_demographics_merged.csv", index=False)
df
```
